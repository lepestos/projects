```python
import math
import random


class Team:
    def __init__(self, n, w, l, rw, rl,wc):
        self.number = n
        self.wins = w
        self.losses = l
        self.roundsWon = rw
        self.roundsLost = rl
        self.roundsDifference = rw - rl
        self.gamesPlayed = w + l
        self.winningCoefficient = wc
    
    def Show(self):
        s = "N" + str(self.number) + "\t"
        s += str(self.wins)  + "\t"
        s += str(self.losses) + "\t"
        s += str(self.roundsWon) + "\t"
        s += str(self.roundsLost) + "\t"
        s += str(self.roundsDifference) + "\t"
        s += str(self.gamesPlayed) + "\t"
        return s
    
    def Update(self):
        self.gamesPlayed = self.wins + self.losses
        self.roundsDifference = self.roundsWon - self.roundsLost
        
        
    def Game(self, team2):
        a = random.randint(1,2)

        if a == 1:
            rw1 = 16
            rw2 = random.randint(0,15)
            
            self.wins += 1
            self.roundsWon += rw1
            self.roundsLost += rw2
            team2.losses += 1
            team2.roundsWon += rw2
            team2.roundsLost += rw1
        if a == 2:
            rw2 = 16
            rw1 = random.randint(0,15)
            
            self.losses += 1
            self.roundsWon += rw1
            self.roundsLost += rw2
            team2.wins += 1
            team2.roundsWon += rw2
            team2.roundsLost += rw1
        s = str(rw1) + " : " + str(rw2)
        #print(s)
        self.Update()
        team2.Update()
        return s
    
    
    def AdvancedGame(self, team2, n):
        rw1 = 0
        rw2 = 0
        sl = 0
        p = self.winningCoefficient**3/(self.winningCoefficient**3 + team2.winningCoefficient**3)
        while (rw1 < 16) and (rw2 < 16):
            sl = random.random()
            if sl <= p:
                rw1 += 1
            else:
                rw2 += 1
        a = 0
        if rw1 == 16:
            a = 1
        else:
            a = 2
        
        if a == 1:
            self.wins += 1
            self.roundsWon += rw1
            self.roundsLost += rw2
            team2.losses += 1
            team2.roundsWon += rw2
            team2.roundsLost += rw1
        if a == 2:
            self.losses += 1
            self.roundsWon += rw1
            self.roundsLost += rw2
            team2.wins += 1
            team2.roundsWon += rw2
            team2.roundsLost += rw1
        s = str(rw1) + " : " + str(rw2)
        self.Update()
        team2.Update()
        return s
        
        
        
def Header():
    s = ""
    s += "#" + "\t"
    s += "Name" + "\t"
    s += "W" + "\t"
    s += "L" + "\t"
    s += "RW" + "\t"
    s += "RL" + "\t"
    s += "RD" + "\t"
    s += "G" + "\t"
    return s

def ShowTable(Teams):
    print(Header())
    for i in range(len(Teams)):
        st = str(i+1) + "\t"
        st += Teams[i].Show()
        print(st)
    print()
    
def Sort(Teams):
    n = len(Teams)
    numeration = []
    l = 0
    for i in range(n):
        numeration.append(i)
        
    for i in range(1,n):
        for j in range(i,0,-1):
            if (Teams[numeration[j]].gamesPlayed >= Teams[numeration[j-1]].gamesPlayed):
                break
            else:
                l = numeration[j]
                numeration[j] = numeration[j-1]
                numeration[j-1] = l
    
    for i in range(1,n):
        for j in range(i,0,-1):
            if (Teams[numeration[j]].roundsLost >= Teams[numeration[j-1]].roundsLost):
                break
            else:
                l = numeration[j]
                numeration[j] = numeration[j-1]
                numeration[j-1] = l
            
    for i in range(1,n):
        for j in range(i,0,-1):
            if (Teams[numeration[j]].roundsWon <= Teams[numeration[j-1]].roundsWon):
                break
            else:
                l = numeration[j]
                numeration[j] = numeration[j-1]
                numeration[j-1] = l
    
    for i in range(1,n):
        for j in range(i,0,-1):
            if (Teams[numeration[j]].wins <= Teams[numeration[j-1]].wins):
                break
            else:
                l = numeration[j]
                numeration[j] = numeration[j-1]
                numeration[j-1] = l
                
    
    table = []
    for i in range(n):
        table.append(Teams[numeration[i]])
    return table

def ScoreGroup(Teams, g, w):
    if w>g:
        return [0,0,0]
    highest = 1
    while (Teams[highest].wins>w):
        highest += 1
    if Teams[highest].wins < w:
        return [0,0,0]
    lowest = highest
    while (Teams[lowest + 1].wins == w):
        lowest += 1
    number = lowest - highest + 1
    
    return [highest + 1, lowest + 1, number]

def Shuffle(Teams):
    n = len(Teams)
    numeration = []
    l = 0
    r = 0
    for i in range(n):
        numeration.append(i)
    for i in range(1,n):
        r = random.randint(0,i)
        for j in range(i,r,-1):
                l = numeration[j]
                numeration[j] = numeration[j-1]
                numeration[j-1] = l
    table = []
    for i in range(n):
        table.append(Teams[numeration[i]])
    return table
    

n = 660
#n = 10
numberOfGames = 16

numberOfSeasons = 1

counter0 = 0
counter1 = 0
counter2 = 0
worstCaseScenario = [0,0,0]
bestCaseScenario = [0,0,0]
worstCaseScenarioPr = 0
bestCaseScenarioPr = 0

for season in range(numberOfSeasons):
    Teams = []

    for i in range(n):
        Teams.append(Team(i+1,0,0,0,0,1-(i+1)/(n+1)))
    Teams = Shuffle(Teams)
    #ShowTable(Teams)    
    nRemaining = n

    for k in range(numberOfGames//4):
        for i in range(nRemaining//20):

            s = Teams[20*i].AdvancedGame(Teams[20*i+9],nRemaining)
            s = Teams[20*i].AdvancedGame(Teams[20*i+19],nRemaining)
            s = Teams[20*i].AdvancedGame(Teams[20*i+6],nRemaining)
            s = Teams[20*i].AdvancedGame(Teams[20*i+16],nRemaining)

            s = Teams[20*i+1].AdvancedGame(Teams[20*i+8],nRemaining)
            s = Teams[20*i+1].AdvancedGame(Teams[20*i+18],nRemaining)
            s = Teams[20*i+1].AdvancedGame(Teams[20*i+5],nRemaining)
            s = Teams[20*i+1].AdvancedGame(Teams[20*i+15],nRemaining)

            s = Teams[20*i+2].AdvancedGame(Teams[20*i+7],nRemaining)
            s = Teams[20*i+2].AdvancedGame(Teams[20*i+17],nRemaining)
            s = Teams[20*i+2].AdvancedGame(Teams[20*i+9],nRemaining)
            s = Teams[20*i+2].AdvancedGame(Teams[20*i+19],nRemaining)

            s = Teams[20*i+3].AdvancedGame(Teams[20*i+6],nRemaining)
            s = Teams[20*i+3].AdvancedGame(Teams[20*i+16],nRemaining)
            s = Teams[20*i+3].AdvancedGame(Teams[20*i+8],nRemaining)
            s = Teams[20*i+3].AdvancedGame(Teams[20*i+18],nRemaining)

            s = Teams[20*i+4].AdvancedGame(Teams[20*i+5],nRemaining)
            s = Teams[20*i+4].AdvancedGame(Teams[20*i+15],nRemaining)
            s = Teams[20*i+4].AdvancedGame(Teams[20*i+7],nRemaining)
            s = Teams[20*i+4].AdvancedGame(Teams[20*i+17],nRemaining)

            s = Teams[20*i+10].AdvancedGame(Teams[20*i+9],nRemaining)
            s = Teams[20*i+10].AdvancedGame(Teams[20*i+19],nRemaining)
            s = Teams[20*i+10].AdvancedGame(Teams[20*i+5],nRemaining)
            s = Teams[20*i+10].AdvancedGame(Teams[20*i+15],nRemaining)

            s = Teams[20*i+11].AdvancedGame(Teams[20*i+8],nRemaining)
            s = Teams[20*i+11].AdvancedGame(Teams[20*i+18],nRemaining)
            s = Teams[20*i+11].AdvancedGame(Teams[20*i+9],nRemaining)
            s = Teams[20*i+11].AdvancedGame(Teams[20*i+19],nRemaining)

            s = Teams[20*i+12].AdvancedGame(Teams[20*i+7],nRemaining)
            s = Teams[20*i+12].AdvancedGame(Teams[20*i+17],nRemaining)
            s = Teams[20*i+12].AdvancedGame(Teams[20*i+8],nRemaining)
            s = Teams[20*i+12].AdvancedGame(Teams[20*i+18],nRemaining)

            s = Teams[20*i+13].AdvancedGame(Teams[20*i+6],nRemaining)
            s = Teams[20*i+13].AdvancedGame(Teams[20*i+16],nRemaining)
            s = Teams[20*i+13].AdvancedGame(Teams[20*i+7],nRemaining)
            s = Teams[20*i+13].AdvancedGame(Teams[20*i+17],nRemaining)

            s = Teams[20*i+14].AdvancedGame(Teams[20*i+5],nRemaining)
            s = Teams[20*i+14].AdvancedGame(Teams[20*i+15],nRemaining)
            s = Teams[20*i+14].AdvancedGame(Teams[20*i+6],nRemaining)
            s = Teams[20*i+14].AdvancedGame(Teams[20*i+16],nRemaining)
        #nRemaining -= 10
        Teams = Sort(Teams)
    counter0 += ScoreGroup(Teams,16,9)[0]
    counter1 += ScoreGroup(Teams,16,9)[1]
    counter2 += ScoreGroup(Teams,16,9)[2]
    
    pr = 100 * (256 - ScoreGroup(Teams,16,9)[0] + 1)/ScoreGroup(Teams,16,9)[2]
    if season == 0:
        worstCaseScenario = ScoreGroup(Teams,16,9)
        bestCaseScenario = ScoreGroup(Teams,16,9)
        worstCaseScenarioPr = pr
        bestCaseScenarioPr = pr
    else:
        if pr < worstCaseScenarioPr:
            worstCaseScenario = ScoreGroup(Teams,16,9)
            worstCaseScenarioPr = pr
        elif pr > bestCaseScenarioPr:
            bestCaseScenario = ScoreGroup(Teams,16,9)
            bestCaseScenarioPr = pr
        
print(counter0/numberOfSeasons)
print(counter1/numberOfSeasons)
print(counter2/numberOfSeasons)
print(worstCaseScenario,", ",worstCaseScenarioPr,"%")
print(bestCaseScenario,", ",bestCaseScenarioPr,"%")

#print(ScoreGroup(Teams,16,9))
ShowTable(Teams)
```
