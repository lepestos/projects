```python
from playsound import playsound as ps
from tkinter import *




def URL(word):
    link = "https://www.macmillandictionary.com/media/british/uk_pron/"
    
    if len(word) >= 5:
        link += word[0]
        link += "/"
        link += word[0:3]
        link += "/"
        link += word[0:5]
        link += "/"
    elif len(word) == 4:
        link += word[0]
        link += "/"
        link += word[0:3]
        link += "/"
        link += word[0:4]
        link += "_/"
    elif len(word) == 3:
        link += word[0]
        link += "/"
        link += word[0:3]
        link += "/"
        link += word[0:3]
        link += "_B/"
    elif len(word) == 2:
        link += word[0]
        link += "/"
        link += word[0:2]
        link += "_/"
        link += word[0:2]
        link += "_Br/"
    elif len(word) == 1:
        link += word[0]
        link += "/"
        link += word[0:1]
        link += "_B/"
        link += word[0:1]
        link += "_Bri/"
        
        
    link += word
    link += "_British_English_pronunciation.mp3"
    return link

#link = "https://www.macmillandictionary.com/media/british/uk_pron/a/amb/ambig/ambiguous_British_English_pronunciation.mp3"
#link2 = "https://www.macmillandictionary.com/media/british/uk_pron/t/tre/treme/tremendous_British_English_pronunciation.mp3"


master = Tk()

def callback(word):
    ps(URL(word))

b = Button(master, text="OK", command=lambda: ps(URL("lion")))
c = Button(master, text="OK", command=lambda: callback("splendid"))
b.pack()
c.pack()
ps(URL("hello"))
mainloop()
```
