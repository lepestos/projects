```python
from matplotlib.pyplot import imshow
import numpy as np
from PIL import Image
import os


def white_space(im_arr,y1,y2,x):
    y = y1
    while (~(im_arr[y,x]==255).all()) & (y < im_arr.shape[0]):
        y += 1
    gap = 0
    while ((im_arr[y,x]==255).all()) & (y < im_arr.shape[0]):
        y += 1
        gap += 1
    return gap
    
    
    
%matplotlib inline

im = Image.open('image.jpg','r')

"""
left = 50
top = 50
right = 165
bottom = 195
im1 = im.crop((left, top, right, bottom))
imshow(np.asarray(im))
im1.save(os.path.join("CroppedImages","im1.jpg"))
"""
imshow(np.asarray(im))
im_arr = np.asarray(im)
print(im_arr[100,200])

y1 = 190
y2 = 230
x = 950

print(white_space(im_arr,y1,y2,x))
```

```python
def label_pixels(row):
    if (row == 255).all():
        return np.array([0,0,0])
    else:
        return np.array([255,255,255])
    
flt = np.apply_along_axis(lambda row: label_pixels(row),axis = 2, arr = im_arr).astype('uint8')


#fltimg = Image.fromarray()
#fltimg.save('filter.jpg')

```
