```python
import time
import numpy as np

def Stat(A,B,k):
    if (k==1):
        if len(B) == 0: return A[k-1]
        if len(A) == 0: return B[k-1]
        if (A[k-1] < B[k-1]):
            return A[k-1]
        else:
            return B[k-1]
    if k % 2 == 0:
        l = k//2
        if A[l-1]<B[l-1]:
            return Stat(A[l:],B[:l],l)
        else:
            return Stat(A[:l],B[l:],l)
    else:
        l = k//2 + 1
        if A[l-1]<B[l-1]:
            return Stat(A[(l-1):],B[:(l-1)],l)
        else:
            return Stat(A[:(l-1)],B[(l-1):],l)
        
#A = [1,2,3,4,5,6,7,8]
#B = [9,10,11,12,13,14,15,16]
A = []
B = []
N = 10000
for i in range(N):
    A.append(2*i)
    B.append(2*i + 1)
#A = [1,2]
#B = [3,4]
Am = np.array(A)
Bm = np.array(B)
k = len(A)

#print(A)
#print(B)

t1 = time.time()
print(Stat(Am,Bm,k))
t2 = time.time()
dt = t2 - t1
print (dt)



A1 = []
B1 = []
N = 10000**2
for i in range(N):
    A1.append(2*i)
    B1.append(2*i + 1)
#A = [1,2]
#B = [3,4]
Am1 = np.array(A1)
Bm1 = np.array(B1)

k1 = len(A1)

#print(A)
#print(B)

t11 = time.time()
print(Stat(Am1,Bm1,k1))
t21 = time.time()
dt1 = t21 - t11
print (dt1)
```
