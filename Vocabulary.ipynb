```python
import random

class Word:
    def __init__(self, lan1, lan2):
        self.lan1 = lan1
        self.lan2 = lan2
    def Check12(self,lan2):
        if self.lan2 == lan2:
            return True
        else:
            return False
    def Check21(self,lan1):
        if self.lan1 == lan1:
            return True
        else:
            return False
    def PrintWord(self):
        s = ""
        s += self.lan1
        s += " - "
        s += self.lan2
        print(s)
    def Is(self,s):
        if s == self.lan1 or s == self.lan2:
            return 1
        else:
            return 0
    def Contains(self,s):
        l = self.lan1.split(" ")
        l.extend(self.lan2.split(" "))
        for el in l:
            if s == el:
                return 1
        return 0
        
    def WordString(self):
        s = ""
        s += self.lan1
        s += " - "
        s += self.lan2
        return s

def RandomPermutation(n):
    r = 0
    lst = []
    for i in range(n):
        lst.append(i)
        r = random.randint(0,i)
        for j in range(r):
            lst[i-j],lst[i-j-1]=lst[i-j-1],lst[i-j]
    return lst

def ConstructList(st,firstString,lastString):
    if lastString == 0:
        lst = st.split("\n")[firstString-1:]
    else:
        lst = st.split("\n")[firstString-1:lastString]
    words = []
    a = ""
    b = ""
    for row in lst:
        a = row.split("-")[0]
        b = row.split("-")[1]
        if a[0] == " ":
            a = a[1:len(a)]
        if a[len(a)-1] == " ":
            a = a[0:(len(a)-1)]
        if b[0] == " ":
            b = b[1:len(b)]
        if b[len(b)-1] == " ":
            a = a[0:(len(b)-1)]
        words.append(Word(a,b))
    return words

def Description():
    print("1: language 1 -> language 2")
    print("2: language 2 -> language 1")
    print("3: learning following 10")
    print("4: language 1 <-> language 2")
    print("5: search")
    print("6: read")
    print("7: random N words 1 -> 2")
    print("8: random N words 2 -> 1")
    print("9: search multiple")

def Mode1(st,firstString,lastString):
    words = ConstructList(st,firstString,lastString)
    n = len(words)
    per = RandomPermutation(n)
    answer = ""
    right = True
    for i in per:
        if (answer == "EXIT"):
            return 1
        right = False
        while(not right):
            answer = input(words[i].lan1 + " -")
            if (answer == "EXIT"):
                return 1
            right = words[i].Check12(answer)
    return 0

def Mode2(st,firstString,lastString):
    words = ConstructList(st,firstString,lastString)
    n = len(words)
    per = RandomPermutation(n)
    answer = ""
    right = True
    for i in per:
        right = False
        while(not right):
            answer = input(words[i].lan2 + " -")
            if (answer == "EXIT"):
                return 1
            right = words[i].Check21(answer)
        if (answer == "EXIT"):
            return 1
    return 0
    
def Mode3(st,firstString):
    repeats = 3
    Mode6(st,firstString,firstString + 4)
    print("...\n...\n...")
    for j in range(repeats):
        exit = Mode1(st,firstString,firstString+4)
        if(exit): return 1
        print("...\n...\n...")
    for j in range(repeats):
        exit = Mode2(st,firstString,firstString+4)
        if(exit): return 1
        print("...\n...\n...")
    Mode6(st,firstString+5,firstString+9)
    print("...\n...\n...")
    for j in range(repeats):
        exit = Mode1(st,firstString+5,firstString+9)
        if(exit): return 1
        print("...\n...\n...")
    for j in range(repeats):
        exit = Mode2(st,firstString+5,firstString+9)
        if(exit): return 1
        print("...\n...\n...")
    exit = Mode1(st,firstString,firstString+9)
    if(exit): return 1
    print("...\n...\n...")
    exit = Mode2(st,firstString,firstString+9)
    if(exit): return 1
    return 0

def Mode4(st,firstString,lastString):
    exit = Mode1(st,firstString,lastString)
    if(exit): return 1
    print("...\n...\n...")
    exit = Mode2(st,firstString,lastString)
    if(exit): return 1
    return 0

def Mode5(st):
    words = ConstructList(st,1,0)
    answer = ""
    exit = 0
    counter = 0
    result = ""
    while exit == 0:
        answer = input("Search: ")
        if answer == "EXIT":
            return 1
        print('-------------')
        counter = 0
        result = ""
        for word in words:
            if word.Contains(answer) or word.Is(answer):
                result += word.WordString()
                result += "\n"
                counter += 1
        if counter == 0:
            print("No matching results")
        elif counter == 1:
            print(counter,end = " matching result: \n")
            print(result)
        else:
            print(counter,end = " matching results: \n")
            print(result)
    return 0

def Mode6(st,firstString,lastString):
    words = ConstructList(st,firstString,lastString)
    for word in words:
        word.PrintWord()
    return 0

def Mode7(st,firstString,lastString,N):
    words = ConstructList(st,firstString,lastString)
    n = len(words)
    per = RandomPermutation(n)[0:N]
    answer = ""
    right = True
    for i in per:
        if (answer == "EXIT"):
            return 1
        right = False
        while(not right):
            answer = input(words[i].lan1 + " -")
            if (answer == "EXIT"):
                return 1
            right = words[i].Check12(answer)
    return 0

def Mode8(st,firstString,lastString,N):
    words = ConstructList(st,firstString,lastString)
    n = len(words)
    per = RandomPermutation(n)[0:N]
    answer = ""
    right = True
    for i in per:
        if (answer == "EXIT"):
            return 1
        right = False
        while(not right):
            answer = input(words[i].lan2 + " -")
            if (answer == "EXIT"):
                return 1
            right = words[i].Check21(answer)
    return 0

def Mode9(st):
    words = ConstructList(st,1,0)
    answer = ""
    exit = 0
    counter = 0
    result = ""
    while exit == 0:
        answer = input("Search: ")
        if answer == "EXIT":
            break
        for word in words:
            if word.Contains(answer) or word.Is(answer):
                result += word.WordString()
                result += "\n"
                counter += 1
    print('-------------')    
    if counter == 0:
            print("/////")
    else:
            print(result)
    return 0

def Mode10(st,firstString,lastString):
    words = ConstructList(st,firstString,lastString)
    n = len(words)
    per = RandomPermutation(n)
    answer = ""
    right = True
    
    c = []
    for i in per:
        master = Tk()
        b = Button(master, text="OK", command=lambda: ps(URL(words[i].lan1)))
        c.append(b)
    for i in per:
        c[i].pack()
    mainloop()
    for i in per:
        if (answer == "EXIT"):
            return 1
        right = False
        while(not right):
            answer = input(words[i].lan1 + " -")
            if (answer == "EXIT"):
                return 1
            right = words[i].Check12(answer)
    
    return 0
    
    
    
    
    
#-------------------------
inp = 1
firstString = 354
lastString = 361
mode = 2
#-------------------------
"""

basic.txt
german.txt
irregularVerbs.txt
R1.txt
norsk.txt
countries.txt

"""
f = open("german.txt", "r", encoding="utf-8")
st = f.read()

if inp == 1:
    Description()
    try: mode = int(input("Enter mode: "))
    except: mode = 0
    if mode == 1 or mode == 2 or mode == 4 or mode == 6 or mode == 7 or mode == 8 or mode == 10:
        firstString = int(input("Input the first string: "))
        lastString = int(input("Input the last string: "))
        if mode == 7 or mode == 8:
            N = int(input("Input the number of words: "))
        print("-------------\n\n\n-------------")
    if mode == 3:
        firstString = int(input("Input the first string: "))
        print("-------------\n\n\n-------------")
        

if mode == 1:
    Mode1(st,firstString,lastString)
elif mode == 2:
    Mode2(st,firstString,lastString)
elif mode == 3:
    Mode3(st,firstString)
elif mode == 4:
    Mode4(st,firstString,lastString)
elif mode == 5:
    Mode5(st)
elif mode == 6:
    Mode6(st,firstString,lastString)
elif mode == 7:
    Mode7(st,firstString,lastString,N)
elif mode == 8:
    Mode8(st,firstString,lastString,N)
elif mode == 9:
    Mode9(st)
elif mode == 10:
    Mode10(st,firstString,lastString)
```
